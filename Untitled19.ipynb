```python
import pandas as pd
cols = [1, 5, 8, 10, 11, 12, 13, 14, 25, 26, 27, 29, 35, 58, 71, 73, 84,98,99, 100, 102, 103, 105, 106, 108]
df = pd.read_excel("/Users/suhruth/Desktop/globalterrorism.xlsx",usecols= cols)
df.head()
```

```python
df.columns
```

```python
df.shape
```

```python
df = df.drop(df[df.iyear == 0].index)
df = df[pd.notnull(df.latitude)]
df = df[pd.notnull(df.longitude)]
df = df[pd.notnull(df.country_txt)]
df = df[pd.notnull(df.city)]
df = df[pd.notnull(df.provstate)]
```

```python
float_cols = [c for c in df.select_dtypes(include=[float]).columns.tolist()]
df[float_cols] = df[float_cols].fillna(0).astype(int)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
str_cols = df.select_dtypes(exclude=[float]).columns.tolist()
df[str_cols] = df[str_cols].fillna('Unknown')
df[str_cols] = df[str_cols].replace(r'^\.*$', 'Unknown', regex=True)
```

```python
df.isnull().sum()
```

```python
import plotly.graph_objects as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
data = dict(type = 'choropleth',
           locations = df_countries3['countries'],zmin=0, zmax=100,
           colorscale = 'Rainbow',
           locationmode = 'country names',
           z = df_countries3['nkill'],
           text = df_countries3['countries'],
           colorbar = {'title' : 'countries by avg killed per attack'})

layout = dict(title = 'Countries by avg killed',
             geo = dict(showframe= False,
                       projection = {'type' : 'mercator'}))

choromap = go.Figure(data=[data], layout=layout)
iplot(choromap)
```

```python
trace = go.Scattergeo(
    geo='geo3',
    lon=df[(df['iyear']>=2000) & (df['nkill']>=7)]['longitude'],
    lat=df[(df['iyear']>=2000) & (df['nkill']>=7)]['latitude'],
    mode='markers',
    marker=go.Marker(
        size = 3,
        opacity=0.7,
        color='#A60000',
    ),
    text=df[(df['iyear']>=2000) & (df['nkill']>=7)]['country_txt']
)

data=[trace]

layout = go.Layout(
    title='Terrorist attacks with 7 or more fatalities in 2015',
    height=700,
    dragmode='zoom',
    geo3=dict(
        projection=dict(
            type='orthographic', 
        ),
        scope='world', 
        showlakes=True,
        showocean=True,
        showland=True,
        showcountries=True,
    )
)

fig = go.Figure(data=data, layout=layout)
iplot(fig)
```
